# Configuração de Web App Manifest para PWA
Este notebook documenta a criação de `manifest.json`, a adição de `theme-color` no HTML e a geração de ícones PWA.

## 1. Configurar manifest.json para PWA
Crie `manifest.json` com propriedades como `name`, `short_name`, `start_url`, `display`, `background_color`, `theme_color` e `icons`.

```python
import json
manifest = {
    "name": "Busca CEP",
    "short_name": "CEP",
    "start_url": "/",
    "display": "standalone",
    "background_color": "#ffffff",
    "theme_color": "#4fc3f7",
    "icons": [
        {"src": "/icons/icon-192x192.png", "sizes": "192x192", "type": "image/png"},
        {"src": "/icons/icon-512x512.png", "sizes": "512x512", "type": "image/png"}
    ]
}
with open('manifest.json', 'w', encoding='utf-8') as f:
    json.dump(manifest, f, indent=2, ensure_ascii=False)
print('manifest.json criado.')
```

## 2. Adicionar link rel="manifest" no HTML
Insira `<link rel="manifest" href="/manifest.json">` no `<head>` do HTML para referenciar o manifesto.

## 3. Definir meta theme-color no HTML
Adicione `<meta name="theme-color" content="#4fc3f7">` para personalizar a barra de ferramentas em navegadores móveis.

## 4. Gerar e incluir ícones no manifesto
Crie os ícones nos tamanhos recomendados e referencie-os no `manifest.json` para instalação como aplicativo.

```python
from PIL import Image
sizes = [(192, 'icons/icon-192x192.png'), (512, 'icons/icon-512x512.png')]
for size, path in sizes:
    img = Image.new('RGBA', (size, size), (79, 195, 247, 255))
    img.save(path)
    print(f'Criado {path}')
```

## 5. Testar e validar o manifesto no navegador
Use ferramentas de desenvolvedor para verificar o manifesto e testar o comportamento PWA. Abra o Application > Manifest no Chrome DevTools.
